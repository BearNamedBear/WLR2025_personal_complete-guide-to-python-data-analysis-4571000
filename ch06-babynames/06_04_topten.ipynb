# 06_04: A yearly top ten

```python
import math
import collections
import dataclasses
import datetime

import numpy as np
import pandas as pd
import matplotlib.pyplot as pp
```

In this video, we will create yearly top tens for male and female names.

We load the data, and this time we index it slightly differently: by sex and year only, and we'll keep the names as data.

```python
allyears = pd.read_feather('allyears.feather')
```

```python
allyears_byyear = allyears.set_index(['sex', 'year']).sort_index()
```

We now build up our query by chaining pd methods. Getting boys in 2000 is a simple matter of a multiindex `.loc`; then we get the most popular names by sorting the df on number, in descending order.

```python
allyears_byyear.loc['M', 2000].sort_values('number', ascending=False)
```

`head(10)` then gives us the top ten.

```python
allyears_byyear.loc['M', 2023].sort_values('number', ascending=False).head(10)
```

In 2023, Liam was king, followed closely by Noah. How about girls? Olivia and Emma

```python
allyears_byyear.loc['F', 2023].sort_values('number', ascending=False).head(10)
```

If we wish to build a table of the names only over multiple years, we should get rid of the index, and take the name only.

```python
allyears_byyear.loc['F', 2023].sort_values('number', ascending=False).head(10).reset_index().name
```

I will make this into a generic function.

```python
def getyear(sex, year):
    return (allyears_byyear.loc[sex, year].sort_values('number', ascending=False).head(10).reset_index().name)
```

To form a table, we collect the series as the columns of a dataframe, labeled by year. We see that Liam took over from Noah, but Noah kept a decent second place. Oliver rose rapidly, while William is losing favor.

```python
pd.DataFrame({year: getyear('M', year) for year in range(2014, 2024)})
```

For females, Olivia has overtaken Emma. Maybe Charlotte is on its way up!

```python
pd.DataFrame({year: getyear('F', year) for year in range(2014, 2024)})
```

Let's take the 2023 top ten, and see them across the history of the database. I first need to reindex the DataFrame to grab individual names more easily.

```python
allyears_byname = allyears.set_index(['sex', 'name', 'year']).sort_index()
```

Interesting, all the top female names have surged quite recently, except for Evelyn, and to some extend Emma and Charlotte.

```python
pp.figure(figsize=(9,2))

for name in getyear('F', 2023):
    pp.plot(allyears_byname.loc['F', name], label=name)

pp.axis(xmin=1880, xmax=2024)
pp.legend(loc='upper left', fontsize=8);
```

Looking at males, William and James are classics, but the rest are recent discoveries.

```python
pp.figure(figsize=(9,2))

for name in getyear('M', 2023):
    pp.plot(allyears_byname.loc['M', name], label=name)

pp.axis(xmin=1880, xmax=2024)
pp.legend(loc='upper left', fontsize=8);
```

How about all time favorites? We get there quickly by selecting females, grouping by name, summing the numbers, sorting by value. I will use the chaining style.

```python
alltime_f = allyears_byyear.loc['F'] \
                            .groupby('name') \
                            .sum() \
                            .sort_values('number', ascending=False) \
                            .head(10)
```

```python
alltime_f
```

Compare with the modest totals of the 2023 top ten. We'll see if they have staying power!

```python
allyears_byname.loc[pd.IndexSlice['F', getyear('F', 2023), :]].groupby('name').sum()
```

Most of the overall top-ten names gained their spots in the first half of the 20th century, and lasted through decades.

```python
pp.figure(figsize=(9,2))

for name in alltime_f.index:
    pp.plot(allyears_byname.loc['M', name], label=name)

pp.axis(xmin=1880, xmax=2024)
pp.legend(loc='upper left', fontsize=8);
```
